```python
import numpy as np
import re

def check_nan(val):
    if np.isnan(val):
        return ''
    else:
        return val

def clean_text(string):
    word = re.sub(r'[^\w\s]', ' ', string)
    return word.strip()
    
def make_query(tuples):
    name1 = (tuples[2])
    name2=(tuples[3])
    email=(tuples[10])
    street = (tuples[4])
    ort = (tuples[7])
    
    name1=clean_text(name1)
    name2=clean_text(name2)
    email=clean_text(email)
    ort = clean_text(ort)
    street=clean_text(street)
    
    query_dict={}
    query_dict[name1]=[]
    query_dict[name1].append(name1.strip())
    if name2 != 'None':
        query_dict[name1].append((name1+' '+name2).strip())
    else:
        query_dict[name1].append(name1)
    if ort != 'None':
        query_dict[name1].append((name1+' '+ort).strip())
    else:
        query_dict[name1].append(name1)
    if name2 != 'None' and ort !='None':
        query_dict[name1].append((name1 + ' ' + name2 + ' ' + ort).strip())
    else:
        query_dict[name1].append(name1)
    
    
    return query_dict
```
